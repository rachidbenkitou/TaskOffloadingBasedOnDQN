```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load the CSV file
file_path = 'Sequential_simulation.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataframe to understand its structure
data.head()

# Set the style
sns.set(style="whitegrid")

# 1. Task Successfully Executed
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Generated tasks', y='Tasks successfully executed', hue='Orchestration algorithm')
plt.title('Task Successfully Executed by Generated Tasks and Orchestration Algorithm')
plt.xlabel('Generated Tasks')
plt.ylabel('Tasks Successfully Executed')
plt.legend(title='Orchestration Algorithm')
plt.show()
```

```python
# 2. Average VM CPU Usage %
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Average VM CPU usage (%)', y='Edge devices count', hue='Orchestration algorithm')
plt.title('Average VM CPU Usage by Edge Devices Count and Orchestration Algorithm')
plt.xlabel('Average VM CPU Usage (%)')
plt.ylabel('Edge Devices Count')
plt.legend(title='Orchestration Algorithm')
plt.show()
```

```python
# 3. Average Execution Delay in seconds
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Average execution delay (s)', y='Generated tasks', hue='Orchestration algorithm')
plt.title('Average Execution Delay by Generated Tasks and Orchestration Algorithm')
plt.xlabel('Average Execution Delay (s)')
plt.ylabel('Generated Tasks')
plt.legend(title='Orchestration Algorithm')
plt.show()
```

```python
# 4. Average Waiting Delay in seconds
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Average waiting time (s)', y='Generated tasks', hue='Orchestration algorithm')
plt.title('Average Waiting Delay by Generated Tasks and Orchestration Algorithm')
plt.xlabel('Average Waiting Delay (s)')
plt.ylabel('Generated Tasks')
plt.legend(title='Orchestration Algorithm')
plt.show()
```

```python
# 5. Average Remaining Power
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Average remaining power (%)', y='Edge devices count', hue='Orchestration algorithm')
plt.title('Average Remaining Power by Edge Devices Count and Orchestration Algorithm')
plt.xlabel('Average Remaining Power (%)')
plt.ylabel('Edge Devices Count')
plt.legend(title='Orchestration Algorithm')
plt.show()
```

```python
# 6. Average Cloud Energy Consumption in W / Data Center
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Cloud energy consumption (W)', y='Edge devices count', hue='Orchestration algorithm')
plt.title('Average Cloud Energy Consumption by Edge Devices Count and Orchestration Algorithm')
plt.xlabel('Cloud Energy Consumption (W)')
plt.ylabel('Edge Devices Count')
plt.legend(title='Orchestration Algorithm')
plt.show()
```

```python
# 5. Average Remaining Power (Line Graph)
plt.figure(figsize=(12, 6))
for algorithm in data['Orchestration algorithm'].unique():
    subset = data[data['Orchestration algorithm'] == algorithm]
    plt.plot(subset['Edge devices count'], subset['Average remaining power (%)'], marker='o', label=algorithm)

plt.title('Average Remaining Power by Edge Devices Count and Orchestration Algorithm')
plt.xlabel('Edge Devices Count')
plt.ylabel('Average Remaining Power (%)')
plt.legend(title='Orchestration Algorithm')
plt.grid(True)
plt.show()
```

```python
# 6. Average Cloud Energy Consumption in W / Data Center (Line Graph)
plt.figure(figsize=(12, 6))
for algorithm in data['Orchestration algorithm'].unique():
    subset = data[data['Orchestration algorithm'] == algorithm]
    plt.plot(subset['Edge devices count'], subset['Cloud energy consumption (W)'], marker='o', label=algorithm)

plt.title('Average Cloud Energy Consumption by Edge Devices Count and Orchestration Algorithm')
plt.xlabel('Edge Devices Count')
plt.ylabel('Cloud Energy Consumption (W)')
plt.legend(title='Orchestration Algorithm')
plt.grid(True)
plt.show()
```
